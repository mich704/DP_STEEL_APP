```python
import os
import pandas as pd

path = r'C:\Users\Michal\Desktop\DP-STEEL-AI_v2\storage\ai_models\image_classifier'
image_classifier_CNN_history = os.path.join(path, 'image_classifier_CNN_history.csv')
image_classifier_VGG16_history = os.path.join(path, 'image_classifier_VGG16_history.csv')
```

```python
df_CNN = pd.read_csv(image_classifier_CNN_history)
df_VGG16 = pd.read_csv(image_classifier_VGG16_history)
df_CNN
```

```python
#plot training history
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_CNN['loss'], label='CNN training loss')
plt.plot(df_CNN['val_loss'], label='CNN validation loss')
plt.plot(df_VGG16['loss'], label='VGG16 training loss')
plt.plot(df_VGG16['val_loss'], label='VGG16 validation loss')
plt.legend()
plt.title('Training history of CNN and VGG16 models')
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df_CNN['binary_accuracy'], label='CNN training accuracy')
plt.plot(df_CNN['val_binary_accuracy'], label='CNN validation accuracy')
plt.plot(df_VGG16['binary_accuracy'], label='VGG16 training accuracy')
plt.plot(df_VGG16['val_binary_accuracy'], label='VGG16 validation accuracy')
plt.legend()
plt.title('Accuracy history of CNN and VGG16 models')
```
